# 多重スリットにおける回折・干渉実験の光の強度分布(2次元版)

### 必要なライブラリのインポート

```python
#%matplotlib notebook
import numpy as np
import cmath
import numba
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
from matplotlib.ticker import LinearLocator
```

### 各種設定

```python
# スリットースクリーン間距離
L = 1.0*10**3
# 光の波長
lam = 5.0*10**(-4)
# 波数
k = (2*np.pi)/lam
```

```python
# スクリーンの初期位置
X_0 = 10.0
# スリット間距離
dx = 0.1
dy = 0.1
```

```python
# スリット数(奇数限定)
N = 5
n = (N-1)*0.5
```

```python
x_slit_count, y_slit_count = np.meshgrid(np.arange(-n, n+1), np.arange(-n, n+1))
```

```python
slit_count_array = np.c_[x_slit_count.flatten(), y_slit_count.flatten()]
slit_count_array.shape
```

### numpy.meshgrid()を利用するための格子点の設定

```python
# スクリーン上の格子の分割数
m = 11
# スクリーンのx,y座標の設定
X,Y = np.meshgrid(np.linspace(-X_0, X_0, m), np.linspace(-X_0, X_0, m))
```

```python
screen_coordinates_array = np.c_[X.flatten(), Y.flatten()]
screen_coordinates_array.shape
```

```python
import itertools
coordinates_comb = np.array(list(itertools.product(screen_coordinates_array, slit_count_array)))
```

```python
coordinates_comb.shape
```

### 各スリットからスクリーン上の点までの距離の計算

```python
R = np.sqrt(L**2+(coordinates_comb[:, 0, 0]-coordinates_comb[:, 1, 0]*dx)**2+(coordinates_comb[:, 0, 1]-coordinates_comb[:, 1, 1]*dy)**2)
```

### 波の重ね合わせ（繰り返し処理）

```python
E = np.exp(1j * k * R).reshape(N**2, m**2)
E_sum = np.cumsum(E, axis = 0)
```

### 光強度の計算

```python
Intensity_array = np.reshape(np.abs(E_sum[N**2-1])**2, (m, m))
Intensity_array.shape
```

### 光の強度分布，干渉縞のグラフ化

```python
# https://learntutorials.net/ja/matplotlib/topic/3385/%E3%82%AB%E3%83%A9%E3%83%BC%E3%83%9E%E3%83%83%E3%83%97
fig = plt.figure(
    figsize = (12, 12),
    facecolor = "gray",
    edgecolor = "black",
)
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(X, Y, Intensity_array, rstride=1, cstride=1, cmap='gist_gray')

ax2 = fig.add_subplot(122)
cf = ax2.contourf(X, Y, Intensity_array, 101, vmin=Intensity_array.min(), vmax=Intensity_array.max(), cmap='gist_gray')
cbar = fig.colorbar(cf)
cbar.locator = LinearLocator(numticks = 11)
cbar.update_ticks()

for ax in {ax1, ax2}:
    ax.set_xlabel(r'$x$')
    ax.set_ylabel(r'$y$')

ax1.set_zlabel(r'$I$')
ax2.set_aspect('equal')
```
