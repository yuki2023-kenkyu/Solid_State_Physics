```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter
import datetime
import pytz
```

### 初期値の設定

```python
half_value_layer = 50
mu = np.log(2)/half_value_layer
N_0 = 10**6
```

### 各粒子のインデックスの配列の用意

```python
x_index = range(N_0)
```

### 乱数を用いた各粒子の飛行距離の計算

```python
np.random.seed(10)
eta = np.random.rand(N_0)
x = -np.log(eta)/mu
```

### ヒストグラムの描画

```python
# グラフ化用
fig = plt.figure(
    figsize = (8, 6),
    facecolor = 'white'
)
# グラフの背景色の設定
ax = fig.add_subplot(1, 1, 1)
# グラフ化用
ax.hist(x, color='green', rwidth=1, bins=100)
# ラベルの設定
ax.set_xlabel(r"$Distance$[cm]")
ax.set_ylabel(r"$Number$")
# 凡例の設定
#ax.legend(loc = 'best')
```

```python
time = datetime.datetime.now(pytz.timezone('Asia/Tokyo'))
#(年、月、日付、時、分、秒、マイクロ秒)

#見やすい形に変換
time = time.strftime('%m%d_monte_1101kadai_1.png')
# 図の保存
fig.savefig(time, facecolor = fig.get_facecolor())
```

### 各粒子の飛行距離

```python
# グラフ化用
fig_2 = plt.figure(
    figsize = (8, 6),
    facecolor = 'yellow'
)
# グラフの背景色の設定
ax_2 = fig_2.add_subplot(1, 1, 1)
# グラフ化用
ax_2.plot(x, x_index, '.', label = 'Monte Carlo Simulation', color = 'blue', markersize = 0.5)
# https://www.haya-programming.com/entry/2018/10/11/131716
ax_2.set_facecolor((1,1,1,0))
# ラベルの設定
ax_2.set_xlabel(r"$Distance[cm]$")
ax_2.set_ylabel(r"$Number$")
# 凡例の設定
#ax_2.legend(loc = 'best')
```

```python
time = datetime.datetime.now(pytz.timezone('Asia/Tokyo'))
#(年、月、日付、時、分、秒、マイクロ秒)

#見やすい形に変換
time = time.strftime('%m%d_monte_1101kadai_2.png')
# 図の保存
fig_2.savefig(time, facecolor = fig_2.get_facecolor())
```
