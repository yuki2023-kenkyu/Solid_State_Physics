### 散乱断面積のグラフ補間用の関数

```python
def cross_section_radiation(E, df_xcom):
    # 与えたEの値を超えるdf_xcomの”Energy”の値のインデックスを全て取得しそのうちの0番目のもの
    # http://taustation.com/numpy-where/#1
    index_max = np.where(df_xcom.iloc[:, 0].values > E)[0][0]
    index_min = index_max - 1
    # コンプトン散乱の場合の２点をつなぐ直線の式の変化の割合と切片を求める
    dycom_dx = (df_xcom.iloc[index_max, 2]-df_xcom.iloc[index_min, 2])/(df_xcom.iloc[index_max, 0]-df_xcom.iloc[index_min, 0])
    b_com = df_xcom.iloc[index_max, 2] - dycom_dx*df_xcom.iloc[index_max, 0]
    # 散乱断面積を推定する
    sigma_com = dycom_dx*E + b_com
    # 光電効果による吸収の場合の２点をつなぐ直線の式の変化の割合と切片を求める
    dyabs_dx = (df_xcom.iloc[index_max, 3]-df_xcom.iloc[index_min, 3])/(df_xcom.iloc[index_max, 0]-df_xcom.iloc[index_min, 0])
    b_abs = df_xcom.iloc[index_max, 3] - dyabs_dx*df_xcom.iloc[index_max, 0]
    # 吸収断面積を推定する
    sigma_abs = dyabs_dx*E + b_abs
    # 吸収断面積と散乱断面積の比を求める
    return sigma_abs/(sigma_com + sigma_com)
```

### モンテカルロ法による，クライン-仁科の式に従う散乱角$\theta$を推定する関数（乱数として使用）
#### 散乱角：入射方向（今回はZ軸）とのなす角

```python
def guess_theta(re, alpha):
    # 各成分で0~1までの乱数をn_count個生成
    n_count = 10**5
    t = np.random.uniform(-np.pi, np.pi, n_count)
    #t = np.linspace(-np.pi, np.pi, 3600)
    y_klein = np.random.uniform(0, 1, n_count)
    # クライン-仁科の式
    klein_nishina = 0.5*re**2 *(1/(1+alpha*(1-np.cos(t))))**2 * (1+np.cos(t)**2 + alpha**2 * (1-np.cos(t))**2 / (1+alpha*(1-np.cos(t))))
    # クライン-仁科の式のグラフの下側にあるy_kleinのインデックスを取得
    klein_index_list = np.where(y_klein < klein_nishina)[0]
    # 取得したインデックスに該当するtの値を1つずつ取り出し，新たなリストに格納
    guess_theta_list = [t[i] for i in klein_index_list]
    return guess_theta_list
```

### 放射線源の設定

```python
# 発生させるγ線の数
radiation_source_number = int(10**4)
# 球の半径
radius = 1
# 電子の半径
re = 1
# γ線のエネルギーの初期値[MeV]
E0 = 0.662
alpha = E0/0.511
```

```python
# γ線の発生モデル
np.random.seed(100)
theta = np.arccos(1-2*np.random.rand(radiation_source_number))
r = radius*np.sin(theta)
phi = 2*np.pi*np.random.rand(radiation_source_number)
```

```python
# データフレームに座標を格納
df_radiation_source = pd.DataFrame({
    'x' : np.cos(theta),
    'y' : r * np.sin(phi),
    'z' : r * np.cos(phi)},
    index = np.arange(1, radiation_source_number+1)
)
```

### 遮蔽体の設定

```python
# Csから遮蔽体までの高さ
height = 5
# 遮蔽体の半径
shield_radius = 1000
# 遮蔽体の厚さ
thickness = 200
```

###  放射線源の遮蔽体の計算体系による制限

```python
e_x = (height/df_radiation_source['z'])*df_radiation_source['x']
e_y = (height/df_radiation_source['z'])*df_radiation_source['y']
df_radiation_source['e_square_sum'] = e_x**2 + e_y**2
df_shield_incident = df_radiation_source.loc[(df_radiation_source['e_square_sum'] < shield_radius**2) & (df_radiation_source['z'] > 0), ['x','y','z']].copy()
df_shield_incident.head(10)
```

```python
# グラフの各種設定
fig_2 = plt.figure(
    figsize = (10, 10),
    facecolor = "black",
    edgecolor = "black",
)
# グラフの出力位置の設定
ax_2 = fig_2.add_subplot(111, projection='3d')
# 面の色の設定
ax_2.xaxis.pane.set_facecolor("black")
ax_2.yaxis.pane.set_facecolor("black")
ax_2.zaxis.pane.set_facecolor("black")
# グラフのプロット
ax_2.plot(df_shield_incident.x, df_shield_incident.y, df_shield_incident.z,  ".", label = r'Monte Carlo Simulation', color = 'yellow', markersize = 0.3)
# https://sabopy.com/py/matplotlib-3d-50/#toc8
ax_2.set_box_aspect((1, 1, 1))
```

```python
time = datetime.datetime.now(pytz.timezone('Asia/Tokyo'))
#(年、月、日付、時、分、秒、マイクロ秒)

#見やすい形に変換
time = time.strftime('%m%d_monte_1122kadai_radiation_source.png')
# 図の保存
fig_2.savefig(time, facecolor = fig_2.get_facecolor())
```

### 各粒子の飛行距離の計算

```python
half_value_layer = 10
mu = np.log(2)/half_value_layer
N_0 = int(len(df_shield_incident))
```

```python
eta = np.random.rand(N_0)
df_shield_incident['distance'] = -np.log(eta)/mu
df_shield_incident.head(10)
```

### 吸収・散乱の判定

```python
# 乱数の発生
df_shield_incident['eta_prime'] = np.random.rand(N_0)
# 関数を呼び出し，散乱・吸収断面積の比率を求める（判定の基準）
sigma_rad = cross_section_radiation(E0, df_xcom)
```

```python
df_shield_incident.loc[df_shield_incident['eta_prime'] < sigma_rad, 'flag'] = '吸収'
df_shield_incident.loc[df_shield_incident['eta_prime'] >= sigma_rad, 'flag'] = '散乱'
```

### "散乱"と判定されたもののみデータフレームから抽出し，新たにデータフレームを作成

```python
# 散乱断面積：散乱する確率
df_radiation_scattering = df_shield_incident[df_shield_incident['flag'] == '散乱'].copy()
```

### モンテカルロ法による散乱角の推定を行う関数の呼び出し

```python
guess_theta_list = guess_theta(re, alpha)
```

### クライン-仁科の式による$\theta$の分布の確認

```python
# グラフ化用
fig = plt.figure(
    figsize = (8, 6),
    facecolor = 'white'
)
# グラフの背景色の設定
ax = fig.add_subplot(1, 1, 1)
# グラフ化用
ax.hist(guess_theta_list, color='purple', rwidth=1, bins=100)
# ラベルの設定
ax.set_xlabel(r"Scattering angle $\theta$[rad]")
ax.set_ylabel(r"$Number$")
```

### 乱数による散乱角の決定

```python
df_radiation_scattering['θ'] = [guess_theta_list[np.random.randint(0, len(guess_theta_list)-1)] for i in range(len(df_radiation_scattering))]
```

### エネルギー計算

```python
df_radiation_scattering['energy'] = E0/(1+alpha*(1-np.cos(df_radiation_scattering['θ'])))
df_radiation_scattering['計算体系'] = '体系内'
df_radiation_scattering.head(10)
```

### n個の粒子それぞれのデータフレームの用意

```python
df_number = [df_radiation_scattering.iloc[[i], :].copy().transpose() for i in range(len(df_radiation_scattering))]
```

### 各データフレーム（粒子）について，以下のアルゴリズムに沿ってデータフレームに値を計算・代入

```python
# 試行回数（散乱回数）を格納するためのリスト
count_list = []
for i in range(len(df_number)):
    # 列名を変更
    df_number[i].set_axis(['1回'], axis = 'columns', inplace=True)
    # カラムに粒子番号を追加
    df_number[i].columns.rename(f"No.{df_radiation_scattering.index[i]}", inplace=True)
    # Csの座標を0列目に追加
    df_number[i].insert(0, "0回", 0)
    df_number[i].loc["energy", "0回"] = E0
    # countの初期化
    count = 1
    # 無限ループ
    while True:
        # データフレームに「count回」列を追加，全ての要素を空白にする
        df_number[i].insert(count+1, f"{count+1}回", " ")       
        # x, z座標を計算
        df_number[i].loc['x', f"{count+1}回"] = df_number[i].loc['distance', f"{count}回"]*np.cos(0.5*np.pi - df_number[i].loc['θ', f"{count}回"]) + df_number[i].loc['x', f"{count}回"]
        df_number[i].loc['z', f"{count+1}回"] = df_number[i].loc['distance', f"{count}回"]*np.sin(0.5*np.pi - df_number[i].loc['θ', f"{count}回"]) + df_number[i].loc['z', f"{count}回"]
        # 計算体系内かどうかの条件式
        if((-shield_radius < df_number[i].loc['x', f"{count+1}回"] < shield_radius) & (0 < df_number[i].loc['z', f"{count+1}回"] < thickness+height)):
            # 条件に当てはまれば「体系内」
            df_number[i].loc['計算体系', f"{count+1}回"] = "体系内"
            # 粒子の飛行距離の計算
            df_number[i].loc['distance', f"{count+1}回"] = -np.log(np.random.rand())/mu
            # 散乱・吸収を決める乱数の発生
            df_number[i].loc['eta_prime', f"{count+1}回"] = np.random.rand()
            # 散乱・吸収断面積の比
            sigma_rad = cross_section_radiation(df_number[i].loc['energy', f"{count}回"], df_xcom)
            # 散乱・吸収の判定
            if(df_number[i].loc['eta_prime', f"{count+1}回"] < sigma_rad):
                df_number[i].loc['flag', f"{count+1}回"] = "吸収"
                break # while文の終了
            # 散乱の場合
            else:
                df_number[i].loc['flag', f"{count+1}回"] = "散乱"
                alpha = df_number[i].loc['energy', f"{count}回"]/0.511
                guess_theta_list = guess_theta(re, alpha)
                df_number[i].loc['θ', f"{count+1}回"] = guess_theta_list[np.random.randint(0, len(guess_theta_list)-1)]
                df_number[i].loc['energy', f"{count+1}回"] = df_number[i].loc['energy', f"{count}回"]/(1+alpha*(1-np.cos(df_number[i].loc['θ', f"{count+1}回"])))
                # 次の試行へ
                count += 1
        else:
            df_number[i].loc['計算体系', f"{count+1}回"] = "体系外"
            break # while文の終了
    count_list.append(count)
```

```python
# 散乱回数が最大となる粒子のインデックスを全て取得し，その0番目のものを取り出す
max_index = [i for i, v in enumerate(count_list) if v == max(count_list)][0]
max_num = int(max_index)
```

```python
df_number[max_num]
```

### リスト内包表記によるベクトル成分の計算(https://www.javadrive.jp/python/list/index17.html#section1)

```python
vec_0_x = [np.array([df_number[i].iloc[0, j] for j in range(0, df_number[i].shape[1]-1)]) for i in range(len(df_number))]
vec_0_z = [np.array([df_number[i].iloc[2, j] for j in range(0, df_number[i].shape[1]-1)]) for i in range(len(df_number))]
vec_1_x = [np.array([df_number[i].iloc[0, j] for j in range(1, df_number[i].shape[1])]) - vec_0_x[i] for i in range(len(df_number))]
vec_1_z = [np.array([df_number[i].iloc[2, j] for j in range(1, df_number[i].shape[1])]) - vec_0_z[i] for i in range(len(df_number))]
```

### 最も散乱回数の多い粒子を１つ取り出し，その軌跡をベクトルで描画

```python
# グラフの各種設定
fig_5 = plt.figure(
    figsize = (8, 6),
    facecolor = "white",
    edgecolor = "black",
)
# グラフの出力位置の設定
ax_5 = fig_5.add_subplot(111)
# https://python.atelierkobato.com/quiver/
# ベクトルの描画
ax_5.quiver(vec_0_x[max_num][:], vec_0_z[max_num][:], vec_1_x[max_num][:], vec_1_z[max_num][:], color = "red", angles = 'xy', scale_units = 'xy', scale = 1.0)
# https://python.atelierkobato.com/rectangle/
# 長方形の描画
rec2 = pat.Rectangle(xy = (-shield_radius, height), width = 2*shield_radius, height = thickness, angle = 0, color = "blue", alpha = 0.5)
ax_5.add_patch(rec2)
ax_5.set_xlim(min(df_number[max_num].iloc[0])-1, max(df_number[max_num].iloc[0])+1)
ax_5.set_ylim(min(df_number[max_num].iloc[2])-1, max(df_number[max_num].iloc[2])+1)
# ラベルの設定
ax_5.set_xlabel(r"$x$")
ax_5.set_ylabel(r"$z$")
```

```python
time = datetime.datetime.now(pytz.timezone('Asia/Tokyo'))
#(年、月、日付、時、分、秒、マイクロ秒)

#見やすい形に変換
time = time.strftime('%m%d_monte_1122kadai_num_max.png')
# 図の保存
fig_5.savefig(time, facecolor = fig_5.get_facecolor())
```
