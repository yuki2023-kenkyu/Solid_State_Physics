```python
#%matplotlib notebook
import numpy as np
import pandas as pd
# 高速化用
from numba import njit, f8, i8, b1, void, prange
# 実行時間測定用
import time
import sympy as sy
# グラフ化用
import matplotlib.pyplot as plt
import matplotlib.patches as pat
from mpl_toolkits.mplot3d import axes3d
from matplotlib.ticker import ScalarFormatter
import matplotlib.ticker as ticker
import datetime
import pytz
```

### 放射線源の設定

```python
radiation_source_number = int(10**5)
radius = 1
```

```python
np.random.seed(100)
theta = np.arccos(1-2*np.random.rand(radiation_source_number))
r = np.sin(theta)
phi = 2*np.pi*np.random.rand(radiation_source_number)
```

```python
"""
df_coordinate = pd.DataFrame({
    'x' : np.random.uniform(-radius, radius, radiation_source_number),
    'y' : np.random.uniform(-radius, radius, radiation_source_number),
    'z' : np.random.uniform(0, 0.1*radius, radiation_source_number)},
    index = np.arange(1, radiation_source_number+1)
)
"""
```

```python
df_coordinate = pd.DataFrame({
    'x' : np.cos(theta),
    'y' : r * np.sin(phi),
    'z' : r * np.cos(phi)},
    index = np.arange(1, radiation_source_number+1)
)
```

```python
df_coordinate.head(10)
```

```python
df_coordinate['x**2 + y**2'] = df_coordinate['x']**2 + df_coordinate['y']**2
df_coordinate
```

### https://note.nkmk.me/python-numpy-append/

```python
df_radiation_source = df_coordinate.loc[df_coordinate['x**2 + y**2'] < radius**2, ['x','y','z']].copy()
df_radiation_source
```

```python
# グラフの各種設定
fig = plt.figure(
    figsize = (10, 10),
    facecolor = "black",
    edgecolor = "black",
)
# グラフの出力位置の設定
ax = fig.add_subplot(111, projection='3d')
# 面の色の設定
ax.xaxis.pane.set_facecolor("black")
ax.yaxis.pane.set_facecolor("black")
ax.zaxis.pane.set_facecolor("black")
# グラフのプロット
ax.plot(df_radiation_source.x, df_radiation_source.y, df_radiation_source.z,  ".", label = r'Monte Carlo Simulation', color = 'yellow', markersize = 0.3)
# https://sabopy.com/py/matplotlib-3d-50/#toc8
ax.set_box_aspect((1, 1, 1))
```

### 遮蔽体の設定

```python
height = 5
shield_radius = 1000
thickness = 200
```

```python
e_x = (height/df_radiation_source['z'])*df_radiation_source['x']
e_y = (height/df_radiation_source['z'])*df_radiation_source['y']
df_radiation_source['e_square_sum'] = e_x**2 + e_y**2
df_radiation_source
```

```python
df_shield_incident = df_radiation_source.loc[(df_radiation_source['e_square_sum'] < shield_radius**2) & (df_radiation_source['z'] > 0), ['x','y','z']].copy()
df_shield_incident.head(10)
```

```python
# グラフの各種設定
fig_2 = plt.figure(
    figsize = (10, 10),
    facecolor = "black",
    edgecolor = "black",
)
# グラフの出力位置の設定
ax_2 = fig_2.add_subplot(111, projection='3d')
# 面の色の設定
ax_2.xaxis.pane.set_facecolor("black")
ax_2.yaxis.pane.set_facecolor("black")
ax_2.zaxis.pane.set_facecolor("black")
# グラフのプロット
ax_2.plot(df_shield_incident.x, df_shield_incident.y, df_shield_incident.z,  ".", label = r'Monte Carlo Simulation', color = 'yellow', markersize = 0.3)
# https://sabopy.com/py/matplotlib-3d-50/#toc8
ax_2.set_box_aspect((1, 1, 1))
```

### 各粒子の飛行距離の計算

```python
half_value_layer = 10
mu = np.log(2)/half_value_layer
N_0 = int(len(df_shield_incident))
```

```python
np.random.seed(10)
eta = np.random.rand(N_0)
df_shield_incident['distance'] = -np.log(eta)/mu
df_shield_incident.head(10)
```

### 吸収・散乱の判定

```python
sigma_tot = 1.0
sigma_ab = 0.7
sigma_sc = sigma_tot-sigma_ab
re = 1
E0 = 660
alpha = 0.511
E = E0
```

```python
np.random.seed(1)
df_shield_incident['eta_prime'] = np.random.rand(N_0)
```

```python
df_shield_incident.loc[df_shield_incident['eta_prime'] < (sigma_ab/sigma_tot), 'flag'] = '吸収'
df_shield_incident.loc[df_shield_incident['eta_prime'] >= (sigma_ab/sigma_tot), 'flag'] = '散乱'
```

```python
# 散乱断面積：散乱する確率
# 微分散乱断面積：立体角dσに散乱する確率
df_radiation_scattering = df_shield_incident[df_shield_incident['flag'] == '散乱'].copy()
```

### クライン-仁科の式の計算

```python
N_1 = int(df_radiation_scattering.shape[0])
np.random.seed(101)
theta = np.random.uniform(-np.pi, np.pi, N_1)
klein = 0.5*re**2 *(1/(1+alpha*(1-np.cos(theta))))**2 * (1+np.cos(theta)**2 + alpha**2 * (1-np.cos(theta))**2 / (1+alpha*(1-np.cos(theta))))
```

### 散乱角による微分散乱断面積の分布

```python
# グラフの各種設定
fig_3 = plt.figure(
    figsize = (8, 6),
    facecolor = "white",
    edgecolor = "black",
)
# グラフの出力位置の設定
ax_3 = fig_3.add_subplot(111)
ax_3.plot(theta, klein, ".", color = "green")
```

### 散乱角の決定

```python
df_radiation_scattering['θ'] = np.random.uniform(-1, 1, N_1)
```

### エネルギー計算

```python
beta = alpha*(1-np.cos(df_radiation_scattering['θ']))
#df_radiation_scattering['dσ/dΩ'] = 0.5*re*(1/(1+beta))**2 * (1+np.cos(df_radiation_scattering['θ'])**2 + (beta**2/(1+beta)))
df_radiation_scattering['energy'] = E0/(1+beta)
df_radiation_scattering['計算体系'] = '体系内'
df_radiation_scattering.head(10)
```

### １つの粒子の軌道を追跡しその結果を格納するためのデータフレームの用意

```python
df_test = df_radiation_scattering.iloc[[7], :].copy().transpose()
df_test.set_axis(['0回'], axis = 'columns', inplace=True)
df_test
```

### 粒子軌道の追跡の実施

```python
# countの初期化
count = 0
# 無限ループ
while True:
    # データフレームに「count回」列を追加，全ての要素を空白にする
    df_test[f"{count+1}回"] = ''
    # x, z座標を計算
    df_test.at['x', f"{count+1}回"] = df_test.at['distance', f"{count}回"]*np.cos(df_test.at['θ', f"{count}回"]) + df_test.at['x', f"{count}回"]
    df_test.at['z', f"{count+1}回"] = df_test.at['distance', f"{count}回"]*np.sin(df_test.at['θ', f"{count}回"]) + df_test.at['z', f"{count}回"]
    # 計算体系内かどうかの条件式
    if((-shield_radius < df_test.at['x', f"{count+1}回"] < shield_radius) & (0 < df_test.at['z', f"{count+1}回"] < thickness+height)):
        # 条件に当てはまれば「体系内」
        df_test.at['計算体系', f"{count+1}回"] = "体系内"
        # 粒子の飛行距離の計算
        df_test.at['distance', f"{count+1}回"] = -np.log(np.random.rand())/mu
        # 散乱・吸収を決める乱数の発生
        df_test.at['eta_prime', f"{count+1}回"] = np.random.rand()
        # 散乱・吸収の判定
        if(df_test.at['eta_prime', f"{count+1}回"] < (sigma_ab/sigma_tot)):
            df_test.at['flag', f"{count+1}回"] = "吸収"
            break # while文の終了
        # 散乱の場合
        else:
            df_test.at['flag', f"{count+1}回"] = "散乱"
            #klein = 0.5*re**2 *(1/(1+alpha*(1-np.cos(theta))))**2 * (1+np.cos(theta)**2 + alpha**2 * (1-np.cos(theta))**2 / (1+alpha*(1-np.cos(theta))))
            df_test.at['θ', f"{count+1}回"] = np.random.uniform(-1, 1)
            beta = alpha*(1-np.cos(df_test.at['θ', f"{count+1}回"]))
            df_test.at['energy', f"{count+1}回"] = df_test.at['energy', f"{count}回"]/(1+beta)
            # 次の試行へ
            count += 1
    else:
        df_test.at['計算体系', f"{count+1}回"] = "体系外"
        break # while文の終了
df_test
```

```python
# グラフの各種設定
fig_4 = plt.figure(
    figsize = (8, 6),
    facecolor = "white",
    edgecolor = "black",
)
# グラフの出力位置の設定
ax_4 = fig_4.add_subplot(111)
# https://python.atelierkobato.com/quiver/
# ベクトルの描画
for i in range(1, df_test.shape[1]):
    ax_4.quiver(df_test.loc['x', f"{i-1}回"], df_test.loc['z', f"{i-1}回"], df_test.loc['x', f"{i}回"]-df_test.loc['x', f"{i-1}回"], df_test.loc['z', f"{i}回"]-df_test.loc['z', f"{i-1}回"], color = "red", angles = 'xy', scale_units = 'xy', scale = 1.0)
# https://python.atelierkobato.com/rectangle/
# 長方形の描画
rec1 = pat.Rectangle(xy = (-shield_radius, height), width = 2*shield_radius, height = thickness, angle = 0, color = "blue", alpha = 0.5)
ax_4.add_patch(rec1)
ax_4.set_xlim(0, 40)
ax_4.set_ylim(0, 15)
```
