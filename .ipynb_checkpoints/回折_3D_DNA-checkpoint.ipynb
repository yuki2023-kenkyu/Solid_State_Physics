# 多重スリットにおける回折・干渉実験の光の強度分布(3次元版)

### 必要なライブラリのインポート

```python
#%matplotlib notebook
import numpy as np
import cmath
import numba
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
from matplotlib.ticker import LinearLocator
```

### 各種設定

```python
# スリットースクリーン間距離
L = 1.2*10**3
# 光の波長
lam = 1.0*10**(-6)
# スクリーンの幅
x_0 = 10.0
# x軸方向のスリット間距離
dx = 0.05
# y軸方向のスリット幅
dy = 0.05
# y軸方向のスリット幅
dz = 0.1
# 波数
k = (2*np.pi)/lam
# スクリーン上の各点との距離を格納するリストの用意
R_points = []
```

### numpy.meshgrid()を利用するための格子点の設定

```python
# 格子の分割数
m = 1025
# スクリーンのx,y座標の設定
X,Y = np.meshgrid(np.linspace(-x_0, x_0, m), np.linspace(-x_0, x_0, m))
# 強度分布格納用配列の初期化
I_points = np.zeros([m, m])
```

### 回転角の設定

```python
# 回転角の設定
p_x = 0
p_y = np.pi/4
p_z = np.pi/4
Theta = np.array([p_x, p_y, p_z])
```

### 座標を回転させる関数の定義

```python
def Rotation(Theta, df):
    # 回転行列
    theta_x = Theta[0]
    theta_y = Theta[1]
    theta_z = Theta[2]
    rotation_z = np.array([[np.cos(theta_z), np.sin(theta_z), 0],
                           [-np.sin(theta_z), np.cos(theta_z), 0],
                           [0, 0, 1]])
    rotation_y = np.array([[np.cos(theta_y), 0, -np.sin(theta_y)],
                           [0, 1, 0],
                           [np.sin(theta_y), 0, np.cos(theta_y)]])
    rotation_x = np.array([[1, 0, 0],
                           [0, np.cos(theta_x), np.sin(theta_x)],
                           [0, -np.sin(theta_x), np.cos(theta_x)]])
    Rotation_matrix = rotation_z@rotation_y@rotation_x
    
    arr_data = df.values
    # 行列積の計算
    new_data = Rotation_matrix@arr_data.T
    return new_data
```

### テキストデータの読み込み

```python
data = open("input.pdb", "r")

with open('pdb.txt', 'w') as f:
    lines = data.readlines()
    line_list = []
    for i in range(len(lines)):
        if 'ATOM' in lines[i][0:4]:
            line = lines[i].replace(lines[i][0:31], '').replace(lines[i][54:80], '')
            line_list.append(line)

    for i in range(len(line_list)):
        f.write(line_list[i])

data.close()

pts = np.loadtxt('pdb.txt')

dna_x, dna_y, dna_z = pts.T
```

```python
df_dna_data = pd.DataFrame(
    data = {'x': dx*dna_x, #x座標の指定
            'y': dy*dna_y, #y座標の指定
            'z': dz*dna_z}
)
```

```python
df_dna_data = df_dna_data.head(30)
```

### 二重らせんを三角関数で表現

```python
phi = 3*np.pi/4
psi = np.pi/32
spiral_h = np.pi/4
t = np.arange(0, 2*np.pi+spiral_h, spiral_h)
spiral1_x = np.cos(t)
spiral1_y = np.sin(t)
spiral1_z = t

spiral2_x = np.cos(t + phi)
spiral2_y = np.sin(t + phi)
spiral2_z = t
"""
spiral3_x = np.cos(t + psi)
spiral3_y = np.sin(t + psi)
spiral3_z = t

spiral4_x = np.cos(t + phi + psi)
spiral4_y = np.sin(t + phi + psi)
spiral4_z = t
"""
```

```python
spiral_x = np.append(spiral1_x, [spiral2_x])
spiral_y = np.append(spiral1_y, [spiral2_y])
spiral_z = np.append(spiral1_z, [spiral2_z])
```

```python
df_spiral_data = pd.DataFrame(
    data = {'x': dx*spiral_x, #x座標の指定
            'y': dy*spiral_y, #y座標の指定
            'z': dz*spiral_z}
)
```

### 座標を回転

```python
new_data = Rotation(Theta, df_dna_data)
```

### 座標を抽出しリスト型に変換

```python
# https://note.nkmk.me/python-pandas-numpy-conversion/
slit_coordinate_x = list(new_data[0])
slit_coordinate_y = list(new_data[1])
slit_coordinate_z = list(new_data[2])
```

### 回転後の座標のプロット

```python
fig2 = plt.figure(
    figsize = (12, 12),
    facecolor = "white",
    edgecolor = "black",
)
ax3 = fig2.add_subplot(111, projection='3d')
ax3.scatter(slit_coordinate_x, slit_coordinate_y, slit_coordinate_z, s=100, c=slit_coordinate_z)
#ax3.plot(slit_coordinate_x, slit_coordinate_y, slit_coordinate_z, "-")
# elevはz軸方向から見た仰角，azimはx, y軸方向の方位角を指定
ax3.view_init(elev=60, azim=270)
ax3.set_xlabel(r'$x$')
ax3.set_ylabel(r'$y$')
```

### スリットの座標数

```python
slit_count = len(slit_coordinate_x)
```

### 各スリットからスクリーン上の点までの距離の計算

```python
for i in range(slit_count):
    R = np.sqrt((L-slit_coordinate_z[i]**2)**2+(X-slit_coordinate_x[i])**2 + (Y-slit_coordinate_y[i])**2)
    R_points.append(R)
# リストを配列型に変換
R_points = np.array(R_points)
```

### 波の重ね合わせ（繰り返し処理）

```python
@numba.jit
def calc_I(slit_count, m, k, R_points, I_points):
    for s in range(m):
        count = 0
        for t in range(m):
            # 電場の初期化
            E = 0
            for u in range(slit_count):
                # 電場の計算
                E += cmath.exp(1j*k*R_points[u, s, t])
            # 光強度の計算結果を格納
            I_points[s, count] = np.abs(E)**2
            count += 1
```

```python
calc_I(slit_count, m, k, R_points, I_points)
```

### 光の強度分布，干渉縞のグラフ化

```python
# https://learntutorials.net/ja/matplotlib/topic/3385/%E3%82%AB%E3%83%A9%E3%83%BC%E3%83%9E%E3%83%83%E3%83%97
fig = plt.figure(
    figsize = (12, 12),
    facecolor = "white",
    edgecolor = "black",
)

ax1 = fig.add_subplot(211, projection='3d')
ax1.plot_surface(X, Y, I_points, rstride=1, cstride=1, cmap='jet')

ax2 = fig.add_subplot(212)
# ３次元の等高線を書く
# https://matplotlib.org/2.0.2/examples/color/colormaps_reference.html
# http://taustation.com/matplotlib-pyplot-contour-contourf/#i-3
cf = ax2.contourf(X, Y, I_points, 51, vmin=40, vmax=I_points.max(), alpha=0.5, cmap='gist_yarg')
#ax1.contourf(X, Y, I_points, zdir='z', offset=1000, cmap='jet')
cbar = fig.colorbar(cf)
# 
cbar.locator = LinearLocator(numticks = 11)
cbar.update_ticks()

for ax in {ax1, ax2}:
    ax.set_xlabel(r'$x$')
    ax.set_ylabel(r'$y$')

ax1.set_zlabel(r'$I$')
ax2.set_aspect('equal')
#ax1.set_zlim(40, I_points.max())
```
