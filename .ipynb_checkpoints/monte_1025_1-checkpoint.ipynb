```python
#%matplotlib notebook
import numpy as np
# グラフ化用
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
from matplotlib.ticker import ScalarFormatter
import matplotlib.ticker as ticker
import datetime
import pytz
```

```python
n_count = 10**5
```

```python
np.random.seed(100)
theta = np.arccos(1-2*np.random.rand(n_count))
r = np.sin(theta)
eta = 2*np.pi*np.random.rand(n_count)
```

```python
x = np.cos(theta)
y = r * np.sin(eta)
z = r * np.cos(eta)
```

```python
# グラフの各種設定
fig = plt.figure(
    figsize = (10, 10),
    facecolor = "black",
    edgecolor = "black",
)
# グラフの出力位置の設定
ax = fig.add_subplot(111, projection='3d')
# 面の色の設定
ax.xaxis.pane.set_facecolor("black")
ax.yaxis.pane.set_facecolor("black")
ax.zaxis.pane.set_facecolor("black")
"""
# 軸目盛の非表示
ax.xaxis.set_major_locator(ticker.NullLocator())
ax.xaxis.set_minor_locator(ticker.NullLocator())
ax.yaxis.set_major_locator(ticker.NullLocator())
ax.yaxis.set_minor_locator(ticker.NullLocator())
ax.zaxis.set_major_locator(ticker.NullLocator())
ax.zaxis.set_minor_locator(ticker.NullLocator())
"""
# グラフのプロット
ax.plot(x, y, z,  ".", label = r'Monte Carlo Simulation', color = 'yellow', markersize = 0.2)
# https://sabopy.com/py/matplotlib-3d-50/#toc8
ax.set_box_aspect((1, 1, 1))
```

```python
time = datetime.datetime.now(pytz.timezone('Asia/Tokyo'))
#(年、月、日付、時、分、秒、マイクロ秒)

#見やすい形に変換
time = time.strftime('%m%d_monte_1025kadai_1.png')
# 図の保存
fig.savefig(time, facecolor = fig.get_facecolor())
```
