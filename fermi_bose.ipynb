```python
# モジュールのインポート
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 初期条件
fermi_0 = -5000
fermi_max = 5000
bose_0 =  0
bose_max = 100
kB = 1
mu = 0
T_points = np.array([[100], [500], [1000]])
```

```python
# グラフ化用リスト
fermi_points = np.arange(fermi_0, fermi_max, 0.01)
bose_points = np.arange(bose_0, bose_max, 0.01)
n_fe = np.zeros([len(T_points),len(fermi_points)])
n_bo = np.zeros([len(T_points),len(bose_points)])
```

```python
# 分布関数
# フェルミ分布
def f_fe(eps, T):
    return 1/(np.exp((eps-mu)/(kB*T))+1)
# ボース分布
def f_bo(eps, T):
    return 1/(np.exp((eps-mu)/(kB*T))-1)
```

```python
# 繰り返し計算
j = 0
while j<len(fermi_points):
    for eps in fermi_points:
        for i in range(len(T_points)):
            n_fe[i][j] = f_fe(eps, T_points[i])
        j += 1
```

```python
# 繰り返し計算
j = 0
while j<len(bose_points):
    for eps in bose_points:
        for i in range(len(T_points)):
            n_bo[i][j] = f_bo(eps, T_points[i])
        j += 1
```

```python
# グラフ化
# グラフの各種設定
fig = plt.figure(
    figsize = (8, 8),
    facecolor = "white",
    edgecolor = "black",
)
# グラフの出力位置の設定
ax = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)
ax.plot(fermi_points, n_fe[0,:], '-', label = 'T = 100', color = 'blue')
ax.plot(fermi_points, n_fe[1,:], '--', label = 'T = 500', color = 'blue')
ax.plot(fermi_points, n_fe[2,:], '-.', label = 'T = 1000', color = 'blue')

ax2.plot(bose_points, n_bo[0,:], '-', label = 'T = 100', color = 'red')
ax2.plot(bose_points, n_bo[1,:], '--', label = 'T = 500', color = 'red')
ax2.plot(bose_points, n_bo[2,:], '-.', label = 'T = 1000', color = 'red')
# ラベルの設定
ax.set_xlabel("$ϵ/k_B$")
ax.set_ylabel("Fermi")
ax2.set_xlabel("$ϵ/k_B$")
ax2.set_ylabel("Bose")

ax2.set_xlim(0, 5)
ax2.set_ylim(0, 1000)
# 凡例の設定
ax.legend(loc = 'best')
ax2.legend(loc = 'best')
```
